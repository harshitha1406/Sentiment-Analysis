```python
import requests
from bs4 import BeautifulSoup
```

```python
search_query="laptops"
base_url="https://www.amazon.in/s?k="
url=base_url+search_query
url
```

```python
header = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36'}
```

```python
search_response=requests.get(url,headers=header)
print(search_response.status_code)
```

```python
cookie={} # insert request cookies within{}
def Search(search_query):
  url="https://www.amazon.in/s?k="+search_query
  print(url)
  page=requests.get(url,cookies=cookie,headers=header)
  if page.status_code==200:
    return page
  else:
    return "Error"
```

```python
def asin(asin):
  url="https://www.amazon.in/dp/"+asin
  print(url)
  page=requests.get(url,cookies=cookie,headers=header)
  if page.status_code==200:
    return page
  else:
    return "Error"
```

```python
def reviews(review_link):
  url="https://www.amazon.in"+review_link
  print(url)
  page=requests.get(url,cookies=cookie,headers=header)
  if page.status_code==200:
    return page
  else:
    return "Error"
```

```python
product_names=[]
response=Search('laptops')
soup=BeautifulSoup(response.content)
for i in soup.findAll('span',{'class':'a-size-medium a-color-base a-text-normal'}):
  product_names.append(i.text)
```

```python
product_names
```

```python
data_asin=[]
response=Search('laptops')
soup=BeautifulSoup(response.content)
for i in soup.findAll("div",{'class':'s-result-item s-asin sg-col-0-of-12 sg-col-16-of-20 sg-col sg-col-12-of-16'}):
  data_asin.append(i['data-asin'])
```

```python
data_asin
```

```python
link=[]
for i in range(len(data_asin)):
  response=asin(data_asin[i])
  soup=BeautifulSoup(response.content)
  for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
    link.append(i['href'])
```

```python
reviews_list = []
for j in range(len(link)):
  for k in range(10):
    response=reviews(link[j]+'&pageNumber='+str(k))
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("span",{'data-hook':"review-body"}):
      reviews_list.append(i.text)
```
